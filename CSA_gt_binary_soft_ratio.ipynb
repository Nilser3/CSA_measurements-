```python
import numpy as np
import pandas as pd
import nibabel as nib
import matplotlib.pyplot as plt
from skimage.measure import perimeter
```

```python
df=[]
path_database = '/mnt/d/crmbm/data_7T_multiclass/masks/'
for sub in range (1,79):
    
    image_gt_gm = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_GM-manual.nii.gz')
    header_info_gm = image_gt_gm.header
    res=float(header_info_gm["pixdim"][2])
    image_gt_gm = np.array(image_gt_gm.get_fdata())
        
    image_gt_sc = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_SC-manual.nii.gz')
    image_gt_sc = np.array(image_gt_sc.get_fdata())
    
    image_dl_gm = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_7t_multiclass_gm.nii.gz')
    image_dl_gm = (image_dl_gm.get_fdata())
    image_dl_sc = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_7t_multiclass_sc.nii.gz')
    image_dl_sc = np.array(image_dl_sc.get_fdata())
   
    image_soft = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_soft.nii.gz')
    image_soft = np.array(image_soft.get_fdata())
    image_soft_sc=(image_soft[:,:,:,0])
    image_soft_gm=(image_soft[:,:,:,1])
        
    image_unc = nib.load(path_database + 'sub-'+ str(sub).zfill(2)+'_T2star_unc-vox.nii.gz')
    image_unc = np.array(image_unc.get_fdata())
    image_unc_sc=(image_unc[:,:,:,0])
    image_unc_gm=(image_unc[:,:,:,1])
    
    for i in range (image_gt_gm.shape[2]):
        csa_sc_gt=(res**2)*(np.sum(image_gt_sc[:,:,i]))
        csa_gm_gt=(res**2)*(np.sum(image_gt_gm[:,:,i]))
        csa_sc_dl=(res**2)*(np.sum(image_dl_sc[:,:,i]))
        csa_gm_dl=(res**2)*(np.sum(image_dl_gm[:,:,i]))
        csa_sc_soft=(res**2)*(np.sum(image_soft_sc[:,:,i]))
        csa_gm_soft=(res**2)*(np.sum(image_soft_gm[:,:,i]))
        csa_sc_unc=(res**2)*(np.sum(image_unc_sc[:,:,i]))
        csa_gm_unc=(res**2)*(np.sum(image_unc_gm[:,:,i]))
        
        sc_contour_gt = perimeter(image_gt_sc[:,:,i]) * res 
        gm_contour_gt = perimeter(image_gt_gm[:,:,i]) * res 
        ratio_sc = sc_contour_gt/csa_sc_gt
        ratio_gm = gm_contour_gt/csa_gm_gt
        
        d = {'sub': sub, 'csa_sc_gt':csa_sc_gt,'csa_gm_gt':csa_gm_gt,'csa_sc_dl':csa_sc_dl,'csa_gm_dl':csa_gm_dl,
            'csa_sc_soft': csa_sc_soft,'csa_gm_soft': csa_gm_soft,'csa_sc_unc': csa_sc_unc,'csa_gm_unc':csa_gm_unc,
            'sc_contour_gt':sc_contour_gt,'gm_contour_gt':gm_contour_gt,'ratio_sc':ratio_sc,'ratio_gm':ratio_gm}
        df.append(d)
data = pd.DataFrame(data=df)
data
```

```python
data.to_excel(path_database + 'csa_7t.xlsx')
```
